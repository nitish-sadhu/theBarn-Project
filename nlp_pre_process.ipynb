```python
import pymongo
from pymongo import MongoClient
import re
import contractions
import pprint
import nltk
from nltk.stem import SnowballStemmer, WordNetLemmatizer
from nltk.corpus import wordnet
```

```python
client = MongoClient()
```

```python
barn_db = client['barn-database']
```

```python
reviews_col = barn_db['reviews']
```

```python
text_revs = []
for rev in reviews_col.find({}, {'text_review' : 1}):
    text_revs.append(rev['text_review'])
```

```python
len(text_revs)
```

```python
# tokenize
# punct removal
# pos tag
# lemmatize
```

```python
# tokenizing
tokens = []
for rev in text_revs:
    tokens.append(rev.split(" "))
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
```

```python
# punctuation removal
tokens_clean = []
for i in range(len(tokens)):
        tagged = []
        temp = []
        #lemmatized = []
        for j in range(len(tokens[i])):
                temp.append(re.sub(r"[^\w]", "", tokens[i][j]).lower())

        tagged = nltk.pos_tag(temp)

        tokens_clean.append(tagged)
```

```python
tokens_clean[0]
```

```python
def nltk_pos_tagger(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:          
        return None
```

```python
tokens_clean[49]
```

```python
wordnet_tagged = map(lambda x: (x[0], nltk_pos_tagger(x[1])), tokens_clean[5])
```

```python
lemma_tokens = []
for i in range(len(tokens_clean)):
        temp = []
        wordnet_tagged = map(lambda x: (x[0], nltk_pos_tagger(x[1])), tokens_clean[i])
        for word, tag in wordnet_tagged:
                if tag == None:
                        temp.append(word)
                else:
                        temp.append(lemmatizer.lemmatize(word, tag))
        lemma_tokens.append(temp)
```

```python
ids = []
for id in reviews_col.find({}, {'_id':1}):
        ids.append(id['_id'])
```
